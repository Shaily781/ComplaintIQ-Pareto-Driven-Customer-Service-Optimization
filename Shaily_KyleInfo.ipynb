```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
kyleinfo = pd.read_csv('https://raw.githubusercontent.com/Invact-Abhay/DOE/refs/heads/main/ComplaintsTable.csv')
kyleinfo.head()
```

```python
 total_complaints = kyleinfo.groupby('ProductCategory')['ComplaintCount'].sum()
 total_complaints
```

```python
 sorted_complaints = total_complaints.sort_values(ascending=False)
 sorted_complaints
```

```python
cumulative_sum = sorted_complaints.cumsum()
cumulative_sum
```

```python
cumulative_percentage = round(cumulative_sum/sorted_complaints.sum()*100,2)
cumulative_percentage
```

```python
fig, ax1=plt.subplots()
sorted_complaints.plot(kind='bar',color='C0', ax=ax1)
ax1.set_ylabel('ComplaintCount')

ax2=ax1.twinx() # twin to use same axis twice
cumulative_percentage.plot(kind='line',color='C1', ax=ax2, marker='D') #marker for value pointers
ax2.set_ylabel('Cumulative Percentage (%)')
plt.title('Pareto Analysis of Product Complaints')
plt.show()
```

```python
Top4 = kyleinfo.query('ProductCategory in ["Kitchenware","Bedding","Jewelry","Books"]')
Top4
```

```python
sorted_returns1 = Top4.groupby('ComplaintReason')['ComplaintCount'].sum().sort_values(ascending=False)
sorted_returns1
```

```python
 cumulative_reason_percentage = round(sorted_returns1.cumsum()/sorted_returns1.sum()*100,2)
 cumulative_reason_percentage
```

```python
fig, ax1=plt.subplots()
sorted_returns1.plot(kind='bar', color='C0',ax=ax1)
ax1.set_ylabel('Complaints Counts')
ax2=ax1.twinx()
cumulative_reason_percentage.plot(kind='line', color='C1', marker='D')
ax2.set_ylabel('Cumumlative %')
plt.title('Pareto Analysis of the Complaints Reasons')
plt.show()
```
